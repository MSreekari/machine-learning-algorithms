```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])
mean_x = np.mean(x)
mean_y = np.mean(y)
m = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x) ** 2)
b = mean_y - m * mean_x
print("Linear equation: y = {:.2f}x + {:.2f}".format(m, b))
y_pred = m * x + b
plt.scatter(x, y, label = "Original Points")
plt.plot(x, y_pred, label = "Best Fit Line", linewidth = 2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```
